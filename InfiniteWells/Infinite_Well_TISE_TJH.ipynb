```python
#Using the time-independent Schrödinger equation to find the eigenfunctions (wavefunctions) and energy levels of a 
#particle in an infinite potential well. 

#Firstly, import various libraries, which will later be used
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import cmath
```

```python
# define constants

L = 1 # width of the potential well

A = np.sqrt(2 / L) # normalisation constant
```

```python
#solution, the analytic solution to the Schrödinger equation at a given point in an infinite potential well

def solution(n, x):
    """
    Function to calculate the value of the wavefunction at a given point in the infinite potential well
    
    Returns:
        psi : float
              the value of the wavefunction
        
    Parameters:
        n : integer
            the order of the wavefunction
            
        x : float
            the position of the particle in the infinite potential well
    """
    
    psi = A * np.sin((n * np.pi * x) / L) #the solution to the wavefunction in an infinite potential well
    return psi
```

```python
#probability, the probability density function of a particle in an infinite potential well

def probability(n):
    """
    Function to calculate the probability density function of a particle in an infinite potential well
    
    Returns:
        prob : array
               the probability density function
    
    Parameter:
        n : integer
            the order of the wavefunction
    """
    
    prob = [] #create an empty array to store the individual probability density function values
    
    for i in np.arange(0, L, L / pltinc):
        prob = np.append(prob, solution(n, i)**2)
    
    return prob
```

```python
#wavefunction, the wavefunction function across the infinite potential well

def wavefunction(n):
    """
    Function to calculate the wavefunction function for a particle in an infinite potential well
    
    Returns:
        psi : array
              the wavefunction function
    
    Parameter:
        n : integer
            the order of the wavefunction
    """
    
    psi = [] #create an empty array in which to calculate wavefunction values
    
    #for loop to fill the array with values of the wavefunction at specified increments in the range of the 
    #infinite potential well
    for i in np.arange(0, L, L / pltinc):   #consider using something other than arange (will cause problems)
        psi = np.append(psi, solution(n, i)) #call the solution function to find wavefunction values
        
    return psi
```

```python
order = 5 #define the order of the wavefunction for prob density and wavefunction plots
pltinc = 1000      #the number of increments over which the wavefunction will be sampled

x_pos = np.linspace(0, L, pltinc)  #create an array filled with the x-position values of the plot

#set the size of the figures
fig = plt.figure(figsize = (8, 9))

#generate two plots on top of each other with appropriate axes labels
plt.subplot(2, 1, 1)
plt.plot(x_pos, wavefunction(order))
plt.xlabel('Position')
plt.ylabel('Wavefunction')

plt.subplot(2, 1, 2)
plt.plot(x_pos, probability(order))
plt.xlabel('Position')
plt.ylabel('Probability density')

plt.show()
```
